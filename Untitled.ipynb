```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('C:\chirag\WA_Fn-UseC_-Telco-Customer-Churn 0.csv')

# Basic info
print(df.shape)
df.head()

# Check for missing values
print(df.isnull().sum())

# Data types and basic stats
print(df.info())
print(df.describe())

# Churn distribution
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()

# Distribution of tenure
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title('Distribution of Customer Tenure')
plt.show()

# Churn vs. Contract Type
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn by Contract Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# For better visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
sns.countplot(data=df, x='Churn')
plt.title('Customer Churn Count')
plt.show()
```

```python
churn_gender = df.groupby(['gender', 'Churn']).size().reset_index(name='count')
sns.barplot(data=churn_gender, x='gender', y='count', hue='Churn')
plt.title('Churn by Gender')
plt.show()
```

```python
sns.histplot(data=df, x='tenure', bins=30, kde=True)
plt.title('Distribution of Tenure')
plt.show()
```

```python
sns.kdeplot(data=df[df['Churn'] == 'Yes']['tenure'], label='Churned', fill=True)
sns.kdeplot(data=df[df['Churn'] == 'No']['tenure'], label='Retained', fill=True)
plt.title('Tenure Distribution by Churn')
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=df, x='MonthlyCharges', y='TotalCharges', hue='Churn', alpha=0.6)
plt.title('Monthly Charges vs Total Charges')
plt.show()
```

```python
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x='InternetService', hue='Churn')
plt.title('Churn by Internet Service Type')
plt.show()
```

```python
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges by Churn')
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
services = ['OnlineSecurity', 'TechSupport', 'StreamingTV', 'StreamingMovies']
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

for idx, service in enumerate(services):
    ax = axes[idx//2, idx%2]
    sns.countplot(data=df, x=service, hue='Churn', ax=ax)
    ax.set_title(f'{service} vs Churn')

plt.tight_layout()
plt.show()
```
